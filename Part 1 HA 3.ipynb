```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
titanic = pd.read_csv(r'C:\Users\Ирина\Downloads\titanic.csv')
```

```python
titanic['Пол'] = np.where(titanic['Sex'] == 'female', 'Женщины', 'Мужчины')
titanic['For counting'] = 1
titanic.rename(columns={'Pclass':'Класс кают'}, inplace=True)
titanic.head()
```

```python
titanic_numbers = titanic.groupby(['Класс кают','Пол'])['PassengerId'].nunique()
titanic_numbers
```

```python
chart5 = alt.Chart(titanic[['For counting','Класс кают','Пол']]).mark_bar(
).encode(
    
    x = alt.X('Класс кают:N', axis=alt.Axis(labelAngle=0)),
    xOffset='Пол',
    y = alt.Y('sum(For counting):Q').axis(None),
    color = alt.Color('Пол:N').scale(range=['blue', 'green']),
).properties(
    title = ['Распределение пассажиров по классу и полу']
)

text5 = chart5.mark_text(align="center", baseline="bottom"
).encode(
    text=alt.Text('sum(For counting):Q'),
    color = alt.Color().scale(range=['black']) 
)   

(chart5+text5).configure_view(
    stroke=None,
)
```
